```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

report = pd.read_csv("./report.csv")
report
```

```python
num_workers1 = report[report["num_workers"] == 1]
num_workers2 = report[report["num_workers"] == 2]
num_workers4 = report[report["num_workers"] == 4]
num_workers8 = report[report["num_workers"] == 8]

plt.plot(num_workers1["num_files_sorted"], num_workers1["execution_time"], label='1 worker')
plt.plot(num_workers2["num_files_sorted"], num_workers2["execution_time"], label='2 workers')
plt.plot(num_workers4["num_files_sorted"], num_workers4["execution_time"], label='4 workers')
plt.plot(num_workers8["num_files_sorted"], num_workers8["execution_time"], label='8 workers')
plt.legend()
plt.ylabel('Tempo de execução (s)')
plt.xlabel('Número de arquivos ordenados')
plt.show()
```
